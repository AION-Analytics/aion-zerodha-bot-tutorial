# Building a Zerodha/Groww Trading Bot Using AION News-to-Signal

Use AION Analytics News-to-Signal as a stock picker from news India workflow. This notebook shows how to generate news based stock signals NSE traders can map into stocks and broker-linked rules.

```python
%pip install aion-news-to-signal kiteconnect pandas
```

```python
from aion import analyze

headline = "RBI hikes repo rate by 25 bps"
result = analyze(headline)
result
```

```python
result["trade_direction_signals"]
```

```python
sector_to_stocks = {
    "IT": ["TCS", "INFY", "WIPRO"],
    "Banks": ["HDFCBANK", "ICICIBANK", "SBIN"],
    "NBFC": ["BAJFINANCE", "CHOLAFIN", "SHRIRAMFIN"],
    "Realty": ["DLF", "LODHA", "GODREJPROP"],
    "Energy": ["RELIANCE", "ONGC", "BPCL"],
    "Automobile and Auto Components": ["MARUTI", "M&M", "TATAMOTORS"],
    "Consumer Durables": ["VOLTAS", "DIXON", "WHIRLPOOL"],
}

long_sectors = result["trade_direction_signals"]["long"]
short_sectors = result["trade_direction_signals"]["short"]

long_stocks = {sector: sector_to_stocks.get(sector, []) for sector in long_sectors}
short_stocks = {sector: sector_to_stocks.get(sector, []) for sector in short_sectors}

long_stocks, short_stocks
```

```python
print("Pseudo trade idea")
print("Headline:", headline)
print("Long sectors:", long_sectors)
print("Short sectors:", short_sectors)
print("Long candidates:", long_stocks)
print("Short candidates:", short_stocks)

print("\nExample interpretation:")
print("Given RBI rate hike -> short Banks / NBFC / Realty, long IT")
```

# Broker integration notes

- Zerodha users can wire these sector outputs into Kite Connect execution or Streak rules.
- Groww users can use the same signals for watchlists and discretionary execution.
